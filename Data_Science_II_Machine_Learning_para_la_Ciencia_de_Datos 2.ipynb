# Descripción del Dataset: League of Legends Match Dataset 2025

**Licencia:** CC BY 4.0 (Creative Commons Attribution 4.0 International) - Este dataset puede ser compartido, modificado y utilizado siempre que se dé la atribución correspondiente.

**Descripción:** Este dataset contiene información detallada sobre partidas de *League of Legends* recolectadas en 2025. Incluye estadísticas de jugadores, desempeño de equipos y metadatos de las partidas, provenientes del servidor EUN1 (Europe Nordic & East). Con 94 atributos, abarca desde detalles básicos como duración y modo de juego hasta métricas avanzadas como daño infligido, oro ganado, ítems comprados y estadísticas finales de los jugadores (e.g., resistencia mágica, velocidad de movimiento). Es ideal para análisis estadísticos, proyectos de machine learning y estudios sobre esports.

**Origen y recolección:** Los datos fueron obtenidos mediante un script personalizado en Python que consulta la API oficial de Riot Games. El proceso comienza con el PUUID de un jugador y se expande a través de los participantes de las partidas. El script está disponible en: [https://github.com/Blizzeq/league-of-legends-data-collector](https://github.com/Blizzeq/league-of-legends-data-collector).

**Características clave:**
- **Detalles de la partida:** Identificador del juego (`game_id`), inicio (`game_start_utc`), duración (`game_duration`), modo (`game_mode`, e.g., CLASSIC, ARAM), tipo (`game_type`), versión (`game_version`), mapa (`map_id`), plataforma (`platform_id`) y cola (`queue_id`).
- **Estadísticas de jugadores:** Kills, deaths, assists, daño total infligido (`total_damage_dealt`), daño a campeones (`total_damage_dealt_to_champions`), oro ganado (`gold_earned`), ítems comprados (`item0` a `item6`), nivel de maestría (`champion_mastery_level`), y estadísticas finales (e.g., `final_attackDamage`, `final_health`).
- **Datos de equipo:** Posición en el equipo (`team_position`), resultado (`win`), y desempeño colectivo como barones y dragones eliminados.

**Resumen del dataset:**
- **Filas:** Representa participantes individuales en partidas (cada fila es un jugador en un juego específico).
- **Columnas:** 94 atributos que capturan una visión completa del rendimiento y contexto de las partidas.
- **Plataforma:** EUN1.

**Casos de uso:**
- Análisis de rendimiento de jugadores y equipos para identificar patrones de éxito.
- Modelos predictivos para predecir victorias basados en estadísticas como KDA o ítems.
- Estudio de estrategias y comportamientos en el juego competitivo.

**Notas importantes:**
- El uso de este dataset debe cumplir con los términos de servicio y políticas de privacidad de Riot Games.
- Para quienes deseen recolectar más datos con el script, es crucial respetar los límites de la API de Riot Games.

Este dataset ofrece una ventana única al mundo competitivo de *League of Legends* en 2025, permitiendo explorar desde el impacto de los campeones hasta las decisiones tácticas que definen las victorias.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo desde Google Drive
file_path = '/content/drive/MyDrive/Colab Notebooks/leaguedata/league_data.xlsx'  # Ajusta la ruta si está en una subcarpeta
df = pd.read_excel(file_path)

# Mostrar las primeras filas para confirmar
print("Primeras 5 filas del dataset:")
display(df.head())
```

Consulta 1: Tasa de victorias por campeón (Top 10)
Objetivo: Identificar qué campeones tienen el mejor desempeño en términos de victorias.


```python
# Calcular la tasa de victorias por campeón
champion_winrate = df.groupby('champion_name')['win'].mean().sort_values(ascending=False).head(10)

# Gráfico de barras
plt.figure(figsize=(12, 6))
champion_winrate.plot(kind='bar', color='lightgreen')
plt.title('Top 10 campeones por tasa de victorias', fontsize=14)
plt.xlabel('Campeón', fontsize=12)
plt.ylabel('Tasa de victorias', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Mostrar los valores
print("Tasa de victorias por campeón (Top 10):")
display(champion_winrate)
```

Objetivo: Ver cómo las estadísticas de kills, deaths y assists (KDA) afectan las victorias.

```python
# Calcular el promedio de KDA según el resultado
kda_stats = df.groupby('win')[['kills', 'deaths', 'assists']].mean()

# Gráfico de barras
plt.figure(figsize=(10, 6))
kda_stats.plot(kind='bar', color=['#FF9999', '#66B2FF', '#99FF99'], figsize=(10, 6))
plt.title('Promedio de KDA según resultado de la partida', fontsize=14)
plt.xlabel('Resultado (True = Victoria, False = Derrota)', fontsize=12)
plt.ylabel('Promedio', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Estadística')
plt.tight_layout()
plt.show()

# Mostrar los valores
print("Promedio de KDA por resultado:")
display(kda_stats)
```

Consulta 3: Distribución de la duración de partidas por modo
Objetivo: Comparar cuánto duran las partidas según el modo de juego (e.g., CLASSIC, ARAM).

```python
# Histograma por modo de juego
plt.figure(figsize=(12, 6))
for mode in df['game_mode'].unique():
    subset = df[df['game_mode'] == mode]
    plt.hist(subset['game_duration'], bins=20, alpha=0.5, label=mode)
plt.title('Distribución de la duración de partidas por modo', fontsize=14)
plt.xlabel('Duración (segundos)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend(title='Modo de juego')
plt.tight_layout()
plt.show()

# Promedio de duración por modo
duration_by_mode = df.groupby('game_mode')['game_duration'].mean()
print("Duración promedio por modo de juego (en segundos):")
display(duration_by_mode)
```

Consulta 4: Daño infligido vs. Daño recibido por rol
Objetivo: Analizar cómo varía el daño infligido y recibido según el rol del jugador (e.g., SUPPORT, CARRY).

```python
# Gráfico de caja (boxplot) para daño infligido y recibido por rol
plt.figure(figsize=(14, 6))

# Daño infligido
plt.subplot(1, 2, 1)
sns.boxplot(x='role', y='total_damage_dealt', data=df, palette='Set2')
plt.title('Daño infligido por rol', fontsize=14)
plt.xlabel('Rol', fontsize=12)
plt.ylabel('Daño infligido', fontsize=12)
plt.xticks(rotation=45)

# Daño recibido
plt.subplot(1, 2, 2)
sns.boxplot(x='role', y='total_damage_taken', data=df, palette='Set2')
plt.title('Daño recibido por rol', fontsize=14)
plt.xlabel('Rol', fontsize=12)
plt.ylabel('Daño recibido', fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Promedio de daño por rol
damage_by_role = df.groupby('role')[['total_damage_dealt', 'total_damage_taken']].mean()
print("Promedio de daño infligido y recibido por rol:")
display(damage_by_role)
```

Consulta 6: Predicción de victorias y análisis de influencia por rol
Objetivo: Usar un modelo de machine learning (Random Forest) para predecir si un jugador ganará la partida basado en estadísticas como kills, deaths, assists, daño infligido, oro ganado y rol. Luego, analizaremos qué variables (incluyendo el rol) son las más importantes para determinar la victoria.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder

# Cargar el archivo desde Google Drive
file_path = '/content/drive/MyDrive/Colab Notebooks/leaguedata/league_data.xlsx'  # Ajusta la ruta
df = pd.read_excel(file_path)

# Seleccionar características relevantes para la predicción
features = ['kills', 'deaths', 'assists', 'total_damage_dealt', 'total_damage_taken',
            'gold_earned', 'role']  # Puedes agregar más como 'vision_score', 'dragon_kills', etc.
target = 'win'  # Variable objetivo (True/False)

# Preparar los datos
# Convertir 'role' a numérico usando LabelEncoder
le = LabelEncoder()
df['role_encoded'] = le.fit_transform(df['role'])

# Crear el conjunto de datos con las características seleccionadas
X = df[features].copy()
X['role'] = df['role_encoded']  # Reemplazar 'role' textual por el codificado
y = df[target].astype(int)  # Convertir True/False a 1/0

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar un modelo Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Hacer predicciones
y_pred = rf_model.predict(X_test)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo: {accuracy:.2f}")
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred, target_names=['Derrota', 'Victoria']))

# Importancia de las características
feature_importance = pd.DataFrame({
    'Feature': features,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

# Gráfico de importancia de características
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance, palette='viridis')
plt.title('Importancia de las características en la predicción de victorias', fontsize=14)
plt.xlabel('Importancia', fontsize=12)
plt.ylabel('Característica', fontsize=12)
plt.tight_layout()
plt.show()

# Mostrar la tabla de importancia
print("\nImportancia de las características:")
display(feature_importance)

# Análisis adicional: Tasa de victorias por rol
role_winrate = df.groupby('role')['win'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
role_winrate.plot(kind='bar', color='lightcoral')
plt.title('Tasa de victorias por rol', fontsize=14)
plt.xlabel('Rol', fontsize=12)
plt.ylabel('Tasa de victorias', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nTasa de victorias por rol:")
display(role_winrate)
```
